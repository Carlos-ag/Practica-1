```python
import socket
import struct
import time
import random

MCAST_GRP = '224.1.1.1'
MCAST_PORT = 5007
MESSAGE = "Hello, game players!"

sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM, socket.IPPROTO_UDP)
ttl = struct.pack('b', 1)
sock.setsockopt(socket.IPPROTO_IP, socket.IP_MULTICAST_TTL, ttl)


# cambiar por la cantidad de preguntas que se quieran simular
num_questions = 5
players = [("Carlos",0), ("Juan",0), ("Pedro",0), ("Maria",0), ("Ana",0)]



def simulate_question_ranking():
    """
    Simula el ranking de los jugadores en una pregunta
    """


    for i in range(len(players)):
        players[i] = (players[i][0], players[i][1] + random.randint(0,100)) # sumar puntaje aleatorio
    players.sort(key=lambda x: x[1], reverse=True)


    send_message("Question ranking:")
    for i in range(len(players)):
        send_message(f"{players[i][0]}: {players[i][1]}")
    

def send_message(message):
    """
    Envia un mensaje a todos los jugadores
    """
    sock.sendto(message.encode(), (MCAST_GRP, MCAST_PORT))
    print(f"Message sent: {message}")

def simulate_user_disconnection():
    for i in range(len(players)):
        if random.randint(0,5) == 1: # 1/5 de probabilidad de que un jugador se desconecte
            send_message(f"User {players[i][0]} disconnected!")
            players.pop(i)
            break

def main():
    """
    Simula el comportamiento del servidor
    """
    send_message("Game Starts Now!")

    for i in range(num_questions):
        time.sleep(3) # simula el tiempo entre preguntas
        print("\n")
        send_message(f"Question {i+1} starts now!\n")
        time.sleep(3)
        send_message("Question ends!\n")
        simulate_question_ranking()
        simulate_user_disconnection()
    


    # felicitar al ganador

    players.sort(key=lambda x: x[1], reverse=True) # ordenar por puntaje
    send_message(f"Winner: {players[0][0]}")
    print(f"Winner: {players[0][0]}")
    


    send_message("Game Ends!")

    sock.close()

main()
```
